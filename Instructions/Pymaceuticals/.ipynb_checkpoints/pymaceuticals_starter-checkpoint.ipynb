```python
mouse_name = clean.loc[clean['Drug Regimen'] == 'Capomulin']['Mouse ID'].iloc[0]
mouse_name
```

```python
# Generate a line plot of tumor volume vs. time point for a mouse treated with Capomulin

y_values = clean.loc[clean['Mouse ID'] == mouse_name]['Tumor Volume (mm3)']
x_values = clean.loc[clean['Mouse ID'] == mouse_name]['Timepoint']
plt.plot(x_values,y_values)
plt.title("Tumor vol vs Timepoint")
plt.xlabel("Timepoint")
plt.ylabel('Tumor Vol')
```

```python
# Generate a scatter plot of average tumor volume vs. mouse weight for the Capomulin regimen
mouse_vol_weight = clean.loc[clean['Drug Regimen'] == 'Capomulin'].groupby('Mouse ID')['Tumor Volume (mm3)','Weight (g)'].mean()
mouse_vol_weight.plot.scatter(x ='Weight (g)',y ='Tumor Volume (mm3)', s = 50,)

plt.title("Tumor Vol vs Mouse Weight")
plt.xlabel("Weight")
plt.ylabel('Volume')

```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen
from scipy.stats import linregress


y_values = mouse_vol_weight['Tumor Volume (mm3)']
x_values = mouse_vol_weight['Weight (g)']
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
mouse_vol_weight.plot.scatter(x ='Weight (g)',y ='Tumor Volume (mm3)', s = 50,)
plt.plot(x_values,regress_values,"r-")

plt.annotate(line_eq,(16,44),fontsize=15,color="red")
plt.title("Tumor Vol vs Mouse Weight")
plt.xlabel("Weight")
plt.ylabel('Volume')
print(f"The r-squared is: {rvalue**2}")
plt.show()
```
